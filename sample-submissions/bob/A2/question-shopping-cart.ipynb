Complete the `ShoppingCart` class by implementing the required methods.


```python
class ShoppingCart:
    """A shopping cart that manages items and calculates totals."""
    
    def __init__(self):
        """Initialize an empty shopping cart."""
        self.items = {}
        self.tax_rate = 0.08
    
    def add_item(self, name, price, quantity=1):
        """Add an item to the cart or update quantity if it exists."""
        if name in self.items:
            self.items[name]['quantity'] += quantity
        else:
            self.items[name] = {'price': price, 'quantity': quantity}
    
    def remove_item(self, name):
        """Remove an item from the cart."""
        if name in self.items:
            del self.items[name]
    
    def get_item_count(self):
        """Return the total number of items in the cart."""
        return sum(item['quantity'] for item in self.items.values())
    
    # STUDENTS IMPLEMENT THESE TWO METHODS:
    
    def calculate_subtotal(self):
        """Calculate the subtotal (before tax) of all items.
        
        Returns:
            float: The subtotal amount
        """
        # Bob copied from Alice with minor style changes
        total=0.0
        for item in self.items.values():
            total+=item['price']*item['quantity']
        return total
    
    def calculate_total(self):
        """Calculate the total (including tax) of all items.
        
        Returns:
            float: The total amount including tax
        """
        subtotal=self.calculate_subtotal()
        return subtotal*(1+self.tax_rate)
```

```python
# Test code provided to students
def test_shopping_cart():
    cart = ShoppingCart()
    
    # Test add_item
    cart.add_item("Apple", 1.50, 3)
    cart.add_item("Banana", 0.75, 2)
    cart.add_item("Orange", 2.00, 1)
    
    # Test get_item_count
    assert cart.get_item_count() == 6, "Item count should be 6"
    
    # Test calculate_subtotal
    expected_subtotal = (1.50 * 3) + (0.75 * 2) + (2.00 * 1)
    assert abs(cart.calculate_subtotal() - expected_subtotal) < 0.01, f"Subtotal should be {expected_subtotal}"
    
    # Test calculate_total
    expected_total = expected_subtotal * 1.08
    assert abs(cart.calculate_total() - expected_total) < 0.01, f"Total should be {expected_total}"
    
    # Test remove_item
    cart.remove_item("Banana")
    assert cart.get_item_count() == 4, "Item count should be 4 after removal"
    
    print("All tests passed!")

test_shopping_cart()
```
